```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('DailyDelhiClimateTrain.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
figure = px.line(data,x='date',y='meantemp',title='Mean Temperature')
figure.show()
```

```python
figure = px.line(data,x='date',y='humidity',title='Humidity in Delhi')
figure.show()
```

```python
figure = px.line(data,x='date',y='wind_speed',title='Wind Speed in Delhi')
figure.show()
```

```python
figure = px.scatter(data,x='humidity',y='meantemp',
                   trendline='ols',
                   title='relationship between temp and humidity')
figure.show()
```

```python
data['date'] = pd.to_datetime(data['date'],format='%Y-%m-%d')
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
print(data.head())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,10))
plt.title('Temp change in Delhi over Years')
sns.lineplot(data=data,x='month',y='meantemp',hue='year')
plt.show()
```

```python
forecast_data = data.rename(columns={'date':'ds',
                                    'meantemp':'y'})
print(forecast_data)
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model,predictions)
```
